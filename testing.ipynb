# Testing Notebook for Template-API

Provides functions to request the different API-endpoints and test the functionality of the different methods and templates

```python
import requests
import json

BASE_URL = "http://127.0.0.1:8000"

def render_template(template_id, params):
    # Sendet eine Anfrage an den /render Endpoint und gibt das gerenderte Prompt zur√ºck
    url = f"{BASE_URL}/render"
    payload = {
        "template_id": template_id,
        "params": params
    }

    response = requests.post(url, json=payload)

    if response.status_code == 200:
        data = response.json()
        print(f"‚úÖ Template-ID: {data['template_id']}")
        print(f"üß† Gesch√§tzte Tokens: {data['estimated_tokens']}")
        print(f"\nüìÑ Gerenderter Prompt: \n")
        print(data["rendered_prompt"])
    else:
        print(f"‚ùå Fehler ({response.status_code}): {response.text}")
```

```python
response = requests.get(f"{BASE_URL}/templates")

if response.status_code == 200:
    templates = response.json()
    for t in templates:
        print(f"{t['id']} - {t['name']}")
    else: 
        print(f"Fehler beim Abrufen: {response.text}")
```

```python
template_id = "persona"

params = {
    "persona": "UX-Researcher",
    "task": "Fasse die wichtigsten Erkenntnisse aus dem Usability-Test zusammen."
}

render_template(template_id, params)
```
